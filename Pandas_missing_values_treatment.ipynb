# Filling Null Values

## 1. Statistical way of filling null values

- For numeric/continueos feature/column 

    - Mean
            When the Data is Normaly distributed/Gaussian Distributed then we can fill the null values with Mean value of data.
        
    - Median
            When the Data is Skewed then we can fill the null values with Median value of data.
            
- For object/String feature/column

    - Mode
            When the Data is mostly 'object datatype' we can fill the null values with Mode value of data.
            

```python
df.Gender.mode()
```

```python
df.Gender.fillna(df.Gender.mode()[0],inplace = True)
```

```python
df.isnull().sum()
```

```python
df.Team.value_counts()
```

```python
df.Team.mode()
```

```python
df.Team.fillna(df.Team.mode()[0], inplace = True)
```
